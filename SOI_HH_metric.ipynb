#### Loading Modules

```python
import os
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

#### Loading the model

```python
emotor = tf.keras.models.load_model("models/emotor.h5")
```

#### Defining Train and Test directory

```python
train_dir = 'data_copy/train/'
test_dir = 'data_copy/test/'
row = 48
col = 48
classes = len(os.listdir('data/train/'))
```

#### Loading Train and Test data

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   zoom_range=0.3,
                                   horizontal_flip=True)

training_set = train_datagen.flow_from_directory(train_dir,
                                                batch_size=64,
                                                target_size=(48,48),
                                                shuffle=True,
                                                color_mode='grayscale',
                                                class_mode='categorical')

test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(test_dir,
                                                batch_size=64,
                                                target_size=(48,48),
                                                shuffle=True,
                                                color_mode='grayscale',
                                                class_mode='categorical')
```

#### Calculating Accuracies

```python
train_loss, train_accu = emotor.evaluate(training_set)
test_loss, test_accu = emotor.evaluate(test_set)
print("final train accuracy = {:.2f} , validation accuracy = {:.2f}".format(train_accu*100, test_accu*100))
```
